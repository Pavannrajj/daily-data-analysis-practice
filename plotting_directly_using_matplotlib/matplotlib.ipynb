# **Plotting directly with Matplotlib**

```python
import numpy as numpy
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

**Fetching Data**

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
df.set_index('Country',inplace= True)
df.head()
```

```python
df.shape
```

**Line Plot**

```python
years = list(range(1980,2014))
years = map(str,years)
df_years = df[years]

total_immigrants = df_years.sum()
total_immigrants
```

```python
fig,ax = plt.subplots()

total_immigrants.index = total_immigrants.index.map(int)
ax.plot(total_immigrants,
        marker = 's',
        markersize = 5,
        color = 'blue',
        linestyle = 'dotted')
ax.set_title('Immigrants b/w 1980 to 2013')
ax.set_ylabel('Total Immigrants')
ax.set_xlabel('Years')

ax.set_xlim(1975,2015)
plt.grid()
ax.legend(['Immigrants'])
plt.show()
```

**Scatter Plot**

```python
fig,ax = plt.subplots(figsize = (8,4))

years = list(map(int,years))
ax.scatter(years,total_immigrants,
           marker = 'x',
           s = 20,
           color = 'darkblue')

plt.xlabel('years')
plt.ylabel('Immigrants')
plt.title('Immigrants from 1980 to 2013')
plt.grid()

ax.legend(['Immigrants'],loc = 'upper center')
plt.show()
```

**Bar Plot**

```python
df.sort_values(['Total'],ascending=False,axis = 0,inplace = True)
df_top5 = df.head()
df_top5.reset_index(inplace=True)
df_top5
```

```python
label = list(df_top5.Country)
label
```

```python
label[2] = 'UK'
label
```

```python
fig,ax = plt.subplots(figsize=(14,7))

ax.bar(label,df_top5['Total'],label = label)
ax.set_title('Immigration trend of top 5 countries')
ax.set_xlabel('years')
```

```python
df_bottom5 = df.sort_values(['Total'],ascending=True).head(5)
df_bottom5 = df_bottom5.reset_index()
```

```python
fig,ax = plt.subplots(figsize = (14,7))

label = list(df_bottom5['Country'])

ax.bar(label,df_bottom5['Total'],
       label = label,
       color = ['green','blue','red','orange','purple']
       )
ax.legend(loc = 'upper left')
ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')

plt.show()
```

**Histogram**

```python
df = df.reset_index()
```

```python
df_country = df.groupby(['Country'])['2013'].sum().reset_index()

fig, ax = plt.subplots(figsize = (10,4))
count = ax.hist(df_country['2013'],bins = 10)

ax.set_title('New Immigrants in 2013') 
ax.set_xlabel('Number of Immigrants')
ax.set_ylabel('Number of Countries')
ax.set_xticks(list(map(int,count[1])))
ax.legend(['Immigrants'])

plt.show()

```

**Sub-plotting**

```python
fig,axs = plt.subplots(1,2,sharey = True)

axs[0].plot(total_immigrants)
axs[0].set_title("Line plot on immigrants")

axs[1].scatter(total_immigrants.index , total_immigrants)
axs[1].set_title("scatter plot on immigrants")

axs[0].set_ylabel("Number of Immigrants")
fig.suptitle("Subplotting Example",fontsize = 15)
fig.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize = (8,4))

axs1 = fig.add_subplot(1,2,1)
axs1.plot(total_immigrants)
axs1.set_title("Line plot on immigrants")

axs2 = fig.add_subplot(1,2,2)
axs2.barh(total_immigrants.index,total_immigrants)
axs2.set_title("Bar plot on immigrants")

fig.suptitle('Subplotting Example',fontsize = 15)
fig.tight_layout()
fig.show()
```
