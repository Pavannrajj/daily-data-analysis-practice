**Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Fetching Data**

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
df.head()
```

```python
df.shape
```

```python
years = list(map(str,range(1980,2014)))
print(years)
```

**Area Plots**

```python
df.sort_values(['Total'],ascending=False,inplace = True,axis = 0)
df_top5 = df.head()
df_top5
```

```python
df_top5
```

```python
df_top5.set_index('Country',inplace = True)
df_top5
```

```python
df_top5.index.name = None
```

```python
df_top5 = df_top5[years].transpose()
df_top5
```

```python
df_top5.plot(kind = 'area',alpha = 0.40,stacked = False,figsize = (20,10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

**Two types of plotting**

Option 1: Scripting layer (procedural method) - using `matplotlib.pyplot as 'plt'`

Option 2: Artist layer (Object oriented method) - using an `Axes` instance from Matplotlib (preferred)

```python
ax = df_top5.plot(kind = 'area',alpha = 0.35,stacked = False,figsize = (20,10))
ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_xlabel('Years')
ax.set_ylabel('Number of Immigrants')
```

```python
df_bottom5 = df.sort_values(['Total'],ascending = True).head(5)
df_bottom5
```

```python
df_bottom5.set_index('Country',inplace =True)
df_bottom5 = df_bottom5.loc[:,years]
df_bottom5.index.name = None
df_bottom5
```

```python
df_bottom5 = df_bottom5.transpose()
```

```python
df_bottom5.head()
```

```python
df_bottom5.plot(kind = 'area',stacked = True,figsize = (20,10),alpha = 0.45)
plt.title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
ax = df_bottom5.plot(kind = 'area',stacked = False,figsize = (20,10),alpha = 0.55)
ax.set_title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

**Histograms**

```python
df['2013'].head()
```

```python
count,bin_edges = np.histogram(df['2013'])
print(count)
print(bin_edges)
```

```python
df['2013'].plot(kind = 'hist',figsize = (8,5))


# add a title to the histogram
plt.title('Histogram of Immigration from 195 Countries in 2013')
# add y-label
plt.ylabel('Number of Countries')
# add x-label
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
df['2013'].plot(kind = 'hist',xticks = bin_edges,figsize = (8,5))
```

```python
df.set_index('Country',inplace = True)
df.index.name = None
df.head()
```

```python
df2 = df.loc[['Denmark','Norway','Sweden'],years]
```

```python
df2
```

```python
df2 = df2.transpose()
```

```python
count,bin_edges = np.histogram(df2,bins = 15)
xmin = bin_edges[0]-10
xmax = bin_edges[-1]+10
df2.plot(kind = 'hist',
         bins = 15,
         xticks = bin_edges,
         xlim = (xmin,xmax),
         color = ['coral','darkslateblue','mediumseagreen'],
         figsize = (10,6))
```

**Bar Charts**
