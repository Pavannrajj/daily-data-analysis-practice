**Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Fetching Data**

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
df.head()
```

```python
df.shape
```

```python
years = list(map(str,range(1980,2014)))
print(years)
```

**Area Plots**

```python
df.sort_values(['Total'],ascending=False,inplace = True,axis = 0)
df_top5 = df.head()
df_top5
```

```python
df_top5
```

```python
df_top5.set_index('Country',inplace = True)
df_top5
```

```python
df_top5.index.name = None
```

```python
df_top5 = df_top5[years].transpose()
df_top5
```

```python
df_top5.plot(kind = 'area',alpha = 0.40,stacked = False,figsize = (20,10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

**Two types of plotting**

Option 1: Scripting layer (procedural method) - using `matplotlib.pyplot as 'plt'`

Option 2: Artist layer (Object oriented method) - using an `Axes` instance from Matplotlib (preferred)

```python
ax = df_top5.plot(kind = 'area',alpha = 0.35,stacked = False,figsize = (20,10))
ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_xlabel('Years')
ax.set_ylabel('Number of Immigrants')
```

```python
import seaborn as sns
sns.regplot(x = 'year',y = 'total',data = df_tot)
```

```python
x =df_tot['year']
y = df_tot['total']
fit = np.polyfit(x,y,deg = 1)
fit
```

```python
df_tot.plot(kind = 'scatter',x = 'year', y = 'total')

plt.title('Total immigration to Canada from 1980 - 2013 with best fit line  ')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

#plt line for best fit
plt.plot(x,fit[0]*x + fit[1],color = 'red')
plt.annotate('y = {0:.0f}x+{1:.0f}'.format(fit[0],fit[1]),xy = (2000,180000))
plt.show()


'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```

**Bubble Plots**

```python
df_t = df[years].transpose()
df_t.index = map(int,df_t.index)
df_t.index.name = 'Year'
df_t.reset_index(inplace = True)
df_t.head()
```

```python
print(list(df_t.columns))
```

```python
norm_brazil = (df_t['Brazil']-df_t['Brazil'].min())/(df_t['Brazil'].max()-df_t['Brazil'].min())

norm_argentina = (df_t['Argentina']-df_t['Argentina'].min())/(df_t['Argentina'].max()-df_t['Argentina'].min())
```

```python
ax0 = df_t.plot(
    kind = 'scatter',
    x = 'Year',
    y = 'Brazil',
    figsize = (14,8),
    alpha = 0.5,
    color = 'green',
    s = norm_brazil*2000 +10,
    xlim = (1975,2015)

)

ax1 = df_t.plot(
    kind = 'scatter',
    x= 'Year',
    y = 'Argentina',
    alpha = 0.5,
    color = 'blue',
    s = norm_argentina * 2000 + 10,
    ax = ax0
)


ax0.set_ylabel('Number of immigrations')
ax0.set_title('Immigration from Brazil and argentina from 1980 - 2013')
ax0.legend(['Brazil','Argentina'],loc = 'upper left')
```

```python
norm_india = (df_t['India']-df_t['India'].min())/(df_t['India'].max()-df_t['India'].min())

norm_china = (df_t['China']-df_t['China'].min())/(df_t['China'].max()-df_t['China'].min())
```

```python
ax0 = df_t.plot(
    kind = 'scatter',
    x = 'Year',
    y = 'India',
    figsize = (14,8),
    color = 'red',
    s = norm_india*2000+10
)

ax1 = df_t.plot(
    kind = 'scatter',
    x = 'Year',
    y = 'China',
    s = norm_china*2000 + 10,
    color = 'green',
    ax = ax0
)

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigrations from China and India from 1980 to 2013')
ax0.legend(['India','China'],loc = "upper left")


plt.show()
```
