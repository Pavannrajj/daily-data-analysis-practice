# **Introduction to Matplotlib and Line Plots**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
```

```python
df.head()
```

```python
df.set_index('Country',inplace = True)
df.head()
```

```python
df.index.name = None
df.head()
```

```python
years = list(map(str,range(1980,2014)))
print(years)
```

**Visualizing Data using Matplotlib**


Plotting in pandas

```python
haiti =df.loc['Haiti',years]
haiti.plot()
```

```python
df.loc['India',years].plot()
```

```python
haiti
```

```python
haiti.index = haiti.index.map(int)
```

```python
haiti.index
```

```python
#Optional
# # a=list(map(int,haiti.index))
# a
```

```python
haiti.plot(kind ='line')
plt.title('Immigration from haiti')
plt.ylabel('Number of Immigrations')
plt.xlabel('Years')
plt.text(2000,6000,'2010 Earthquake')
plt.show()
```

```python
df2 = df.loc[['China','India'],years]
df2
```

```python
# total = df[years].sum()
# df[total] = total
df[years].sum()
```

```python
df['Total'] = df[years].sum(axis = 1)
df.head()
```

```python
df.sort_values(by = 'Total',ascending=False,inplace = True,axis = 0)
```

```python
df.head()
```

```python
df3 = df[years].head(5)
df3 = df3.transpose()
```

```python
df3.plot()
plt.legend()
plt.title('Immigration of top 5 contries')
plt.ylabel('Number of immigrations')
plt.xlabel('years')
plt.show()
```
